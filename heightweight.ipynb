```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('weight-height.csv')
```

```python
df.head()
```

```python
df.Gender = df['Gender'].replace(['Male', 'Female'], [0,1])
df.head()
```

```python
df.tail()
```

```python
df1=df.copy()
df2=df.copy()
```

```python
x= df1.drop('Weight',axis=1)
```

```python
x.head()
```

```python
y=df1[['Weight']]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.70,random_state=42)
```

```python
xtrain.head()
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(xtest,ytest)
```

```python
reg.predict([[0,73.847017]])
```

```python
df1['predictedWeight']=reg.predict(x)
```

```python
pre= reg.predict(x)
```

```python
reg.score(xtrain,ytrain)
```

```python
df1.head()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y,pre)
mse
```
