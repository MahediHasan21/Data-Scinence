# Data Sampling Technic

# Synthetic Minority Oversampling Technique (SMOTETomek)

```python
!pip install imblearn
```

```python
from imblearn.combine import SMOTETomek
smt=SMOTETomek(random_state=100)
X,Y=smt.fit_resample(x,y)
```

```python
X.shape
```

```python
Y.shape
```

```python
Y.value_counts()
```

# Near_Miss for Under_Sampling

```python
from imblearn.under_sampling import NearMiss
nm=NearMiss()
x2,y2=nm.fit_resample(x,y)
```

```python
x2.shape
```

```python
y2.shape
```

```python
y2.value_counts()
```

# Over Sampling

```python
df2.head()
```

```python
from imblearn.over_sampling import RandomOverSampler
ros=RandomOverSampler(random_state=100)
x_new, y_new=ros.fit_resample(x,y)
```

```python
x_new.shape
```

```python
y_new.shape
```

```python
y_new.value_counts()
```

# Hold Out Cross Validation, (validation) part

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_new,y_new,random_state=42,train_size=0.75)#over sampling data(x_new,y_new)
```

```python
xtrain1,xtest1,ytrain1,ytest1=train_test_split(x,y,random_state=42,train_size=0.75)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
```

```python
dtc.score(xtest,ytest)
```

# without sampling result

```python
xtrain1,xtest1,ytrain1,ytest1=train_test_split(x,y,random_state=42,train_size=0.75)
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(xtrain1,ytrain1)
```

```python
dtc.score(xtest1,ytest1)
```
